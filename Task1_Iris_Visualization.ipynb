# Task 1: Iris Dataset Analysis

## Objective
Explore and visualize the Iris dataset using Python libraries like pandas, seaborn, and matplotlib.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset from seaborn library
df = sns.load_dataset('iris')
 
# Display the first 5 rows of the dataset to get an overview 
df.head()
```

## Data Understanding
The Iris dataset has 150 samples with 4 features:
- `sepal_length`
- `sepal_width`
- `petal_length`
- `petal_width`
- `species` (target column)


```python
# Print the shape (rows, columns) of the dataset
print("Shape:", df.shape)

# Print the column names to understand the features
print("Columns:", df.columns)             

# Display detailed info about dataset including data types and missing values
df.info()

# Show basic statistical summary for numerical columns (mean, std, min, max, etc.)
df.describe()
```

## Scatter Plot
We will plot sepal length vs petal length to see their relationship.


```python
# Scatter plot to visualize relationship between sepal length and petal length
 
plt.scatter(df['sepal_length'], df['petal_length'])
plt.xlabel("Sepal Length")  # X-axis label
plt.ylabel("Petal Length")  # Y-axis label
plt.title("Sepal vs Petal Length")  # Plot title
plt.show()
```

## Histogram
We will check the distribution of sepal length in the dataset.


```python
# Histogram to see the distribution of sepal length values
 
 df['sepal_length'].hist()
plt.title("Distribution of Sepal Length")  # Plot title
plt.xlabel("Sepal Length")  # X-axis label
plt.ylabel("Frequency")  # Y-axis label
plt.show()
```

## Box Plot
Box plot shows the spread of sepal length and highlights any outliers.


```python
# Boxplot to identify spread and outliers in sepal length
 
sns.boxplot(x=df['sepal_length'])
plt.title("Box Plot of Sepal Length")  # Plot title
plt.show()
```

## Conclusion

- Sepal length and petal length show a positive relationship.
- Most values are concentrated in a specific range.
- A few outliers are present in the dataset.
- Visualizations give a clear understanding of data distribution and spread.
